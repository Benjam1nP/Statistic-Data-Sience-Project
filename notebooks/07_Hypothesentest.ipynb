# 07 Hypothesentests
------------------------------------------------------------
## Hypothesen

**1.**  Die Trinkgeld-Rate (Tip Percentage), gemessen als Verhältnis von Trinkgeld zu Gesamtbetrag, wird signifikant durch die
    Tageszeit beeinflusst.

**2.** Nachts sind Taxis schneller unterwegs als tagsüber

**3.**  Es gibt eine Beste Zeit zum Taxifahren.

**4.** Airport-Fahrten haben eine andere Preisverteilung als Stadt-Fahrten
